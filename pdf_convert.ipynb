```python
from flask import Flask, jsonify,send_from_directory,request
from werkzeug.security import safe_join
from werkzeug.utils import secure_filename
import os
import asyncio
from olmocr.olmocr.pipelineFunc import main
import argparse
import json

app = Flask(__name__)
# 设置上传文件夹的路径
UPLOAD_FOLDER = './pdfs'
app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER
def convert(res_path,out_dir):
    res_txt_list =[]
    for file in os.listdir(res_path):
        if file.endswith('.jsonl'):
            input_file = os.path.join(res_path,file)
            output_path = os.path.join(out_dir, file.replace('.jsonl', '.txt'))
            with open(input_file,'r',encoding='utf-8') as inputFile:
                lines =  inputFile.readlines()
            res_list = []
            with open(output_path,'w',encoding='utf-8') as outFile:
                for line in lines:
                    # 解析 JSON 数据
                    data = json.loads(line)
                    # 提取文本字段（假设键为 'text'，根据实际情况修改）
                    text = data['text']
                    res_list.append(text)
                    print(res_list,'444')
                     # 将文本写入输出文件  
                    outFile.write(text)
        res_txt_list.append('\n\n '.join(res_list))
    return res_txt_list
#转单个文件
def convert_Single(res_path,out_dir):
    res_txt_list =[]
    for file in os.listdir(res_path):
        if file.endswith('.jsonl'):
            input_file = os.path.join(res_path,file)
            output_path = os.path.join(out_dir, file.replace('.jsonl', '.txt'))
            with open(input_file,'r',encoding='utf-8') as inputFile:
                lines =  inputFile.readlines()
            res_list = []
            with open(output_path,'w',encoding='utf-8') as outFile:
                for line in lines:
                    # 解析 JSON 数据
                    data = json.loads(line)
                    # 提取文本字段（假设键为 'text'，根据实际情况修改）
                    text = data['text']
                    res_list.append(text)
                    print(res_list,'444')
                     # 将文本写入输出文件  
                    outFile.write(text)
        res_txt_list.append('\n\n'.join(res_list))
    print(res_txt_list,'oooooo')
    return res_txt_list
def isFileExit(workspace_path):
    if os.path.exists(workspace_path) and os.path.isdir(workspace_path):
    # 遍历目录中的所有文件和子目录
        for filename in os.listdir(workspace_path):
            file_path = os.path.join(workspace_path, filename)  # 获取文件的完整路径
            if os.path.isfile(file_path):  # 判断是否为文件
                os.remove(file_path)  # 删除文件
                print(f"已删除文件: {file_path}")
            elif os.path.isdir(file_path):  # 如果是子目录，可以选择递归删除或跳过
                print(f"跳过子目录: {file_path}")
        print("目录中的所有文件已删除。")
    else:
        print(f"路径不存在或不是一个目录: {workspace_path}")
@app.route('/api/upload-pdf', methods=['POST'])
def upload_audio():
    # 获取上传的文件
    # print(request.files) 
    # # uploaded_files = request.files.getlist('pdfFile')
   
    # file = request.files['file']
    # print(file,'00000000000000')
    # filename = secure_filename(file.filename)
    folder_name = request.headers.get('X-Folder-Name')
    if not folder_name:
        return jsonify({"error": "No folder name provided"}), 400
    # 确保文件夹路径安全
    upload_folder = secure_filename(folder_name)
    # 如果文件夹不存在，则创建文件夹
    if not os.path.exists(upload_folder):
        os.makedirs(upload_folder)
    else:
        isFileExit(upload_folder)

    if not request.data:
        return jsonify({"error": "No file content"}), 400
    
    # 获取文件名
    filename = secure_filename(request.headers.get('Content-Disposition', '').split('filename=')[-1].strip('"'))

    if not filename:
        return jsonify({"error": "No filename provided"}), 400
    file_path = os.path.join(upload_folder, filename)
    with open(file_path, 'wb') as f:
        f.write(request.data)
    
    # file.save(file_path)
    # saved_files.append(filename)
    #存储pdf文件
    # for file in uploaded_files:
    #     # 检查文件扩展名是否为 PDF
    #     if file.filename.lower().endswith('.pdf'):
    #         # 保存文件到上传文件夹
    #         filename = secure_filename(file.filename)
    #         file_path = os.path.join(app.config['UPLOAD_FOLDER'], filename)
    #         file.save(file_path)
    #         saved_files.append(filename)
    return jsonify({"message": "Files uploaded successfully", "status":'1'})
@app.route('/api/file/<path:filename>', methods=['GET'])
def read_file(filename):
    
    # 定义工作目录和参数
    workspace_path = "./localwork"
    pdfs_list=[]
    for file in os.listdir(filename):
        pdfs_list.append(str(os.path.join(filename,file)))
    # 解析pdf
    main(workspace_path,pdfs_list)
    # main(args)
    res_path = os.path.join(workspace_path,"results")
    out_path = './txt_results'
    #将jsonl转成txt
    res_txt = convert(res_path,out_path)
    print(res_txt,'ppppppp')
    #下载所有的txt文件
    # download_files(out_path)
    return {'message':'转换成功','allres':res_txt}
@app.route('/api/filetxt/<path:filename>', methods=['GET'])
async def read_single_file(filename):
    # 定义工作目录和参数
    workspace_path = "./localwork"
    workres_path="./localwork/results"
    pdfs_list=[filename]
    print(pdfs_list,'kkkkkkk')
    
    fileDir = filename.split("/")[0]
    # isFileExit(workres_path)
    
    # 解析pdf
    await main(workspace_path,pdfs_list)
    # main(args)
    res_path = os.path.join(workspace_path,"results")
    out_path = './txt_results'
    # isFileExit(out_path)
    #将jsonl转成txt
    res_txt = convert_Single(res_path,out_path)
    print(res_txt,'ppppppp')
    isFileExit(out_path)
    isFileExit(workres_path)
    # #下载所有的txt文件
    # download_files(out_path)
    return {'message':'转换成功','allres':res_txt}
def download_files(filePath):
    for file in os.listdir:
        send_from_directory(filePath, file,as_attachment=True)
        
    #获取文件夹路径
if __name__ == '__main__':
    app.run(host='0.0.0.0', port=8080, debug=False)
```
